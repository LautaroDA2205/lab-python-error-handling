# Step 1: Define the function for initializing the inventory with error handling
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory

````python
productos = ["t-shirt", "mug", "hat", "book", "keychain"]
def inventario_inicial(productos):
    inventario = {}
    for producto in productos:
        cantidad_valida = False
        while not cantidad_valida:
            try:
                cantidades_producto = int(input(f"Ingrese la cantidad inicial de {producto}: "))
                if cantidades_producto < 0:
                    print("Por favor, ingrese un número entero no negativo.")
                else:
                    inventario[producto] = cantidades_producto
                    cantidad_valida = True
            except ValueError:
                print("Ha ocurrido un error. Por favor, ingrese un número entero.")
        print(f"Inventario inicial de {producto}: {inventario[producto]} unidades.")
    return inventario
````

````python
productos = ["t-shirts", "mugs", "hats", "books", "keychains"]
inventario = inventario_inicial(productos)
print("Inventario inicial completo:", inventario)
````

````python
def calcular_precio_total(productos):
    total = 0
    for producto in productos:
        while True:
            try:
                precio = float(input(f"Ingrese el precio de {producto}: "))
                if precio < 0:
                    print("El precio no puede ser negativo.")
                else:
                    total += precio
                    break
            except ValueError:
                print("Por favor, ingrese un precio válido.")
    return total
````

````python
orden_cliente1 = {
    "t-shirts": 2,
    "mugs": 1,
    "hats": 3,
    "books": 1,
    "keychains": 4
}
total_orden1 = calcular_precio_total(orden_cliente1.keys())
print(f"El precio total de la orden del cliente 1 es: ${total_orden1:.2f}")
````

````python
def obtener_orden_cliente (inventario):
    while True:
        try:
            num = int(input("¿Cuántos pedidos desea realizar?: "))
            if num < 0:
                print("Debe ingresar un número no negativo.")
            else:
                break
        except ValueError:
            print("Ingrese un número válido.")
    pedidos = set()
    for _ in range(num):
        while True:
            producto = input("Producto: ")
            try:
                if producto in inventario and inventario[producto] > 0:
                    pedidos.add(producto)
                    break
                else:
                    print("Producto no válido o sin stock.")
            except:
                print("Error. Ingrese un producto válido.")
    return pedidos
````
